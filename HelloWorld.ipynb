# This is a project to plot Hello World using matplotlib and show it on github/nbviewer

###### import python packages

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

##### Plot the words "hello world" using matplotlib annotate

```python
fig, ax = plt.subplots(figsize = (16,2))
ax.annotate(s = 'Hello World', xy = (0,0), xycoords = 'axes fraction', fontsize = 150)

plt.axis('off');
```
