```python
####################################################
# (c) Lars Tijssen
# plot a random walk starting in point (0,0)
####################################################

%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
%%time

# starting position
pos = np.array([0,0])

# steps size
n = 1000

# number of random walks
m = 20

poslist = []

random_dir = np.random.randint(high = 5, low = 0, size = n)

random_x = np.zeros(shape = random_dir.shape)
random_y = np.zeros(shape = random_dir.shape)
```

```python
random_x[random_dir == 0] = 1
random_x[random_dir == 1] = -1
random_x[random_dir == 2] = 2
random_y[random_dir == 3] = -1
random_y[random_dir == 4] = 1
```

```python
x = random_x.reshape(m,n/m).cumsum(axis = 1)
y = random_y.reshape(m,n/m).cumsum(axis = 1)
```

```python
fig, ax = plt.subplots(figsize = (16,9))

for x_m, y_m in zip(x, y):
    ax.plot(x_m, y_m, linewidth = .25)
    
ax.set_xlim(x.min(),x.max())
ax.set_ylim(y.min(),y.max())

ax.grid()
```
